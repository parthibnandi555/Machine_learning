```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs
```

```python
x, y_true = make_blobs(n_samples=500, centers=3, cluster_std=0.1, random_state=0)
```

```python
x
```

```python
df = pd.DataFrame(x, columns=['feature 1', 'feature 2'])
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
inertia= []
k_range = range(1, 11)
```

```python
for k in k_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x_scaled)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(k_range, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt
```

```python
kmean_final = KMeans(n_clusters=2,random_state=42)
kmean_final.fit(x_scaled)
```

```python
cluster_labels =kmean_final.predict(x_scaled)
```

```python
cluster_labels
```

```python
df['cluster']= cluster_labels
```

```python
sns.scatterplot(data=df, x='feature 1', y='feature 2', hue='cluster')
```
